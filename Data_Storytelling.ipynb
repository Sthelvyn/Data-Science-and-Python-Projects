```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Cores
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

AZUL = "#003F5c"
CINZA = "#9E9E9E"
CINZA_TEXTO = "#515151"
```

```python
# dados ficticios
X = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]
alunos = [180, 172, 190, 173, 161, 197, 202, 187, 183, 210, 187, 172]
projetos = [180, 170, 184, 169, 159, 187, 192, 180, 169, 171, 140, 124]
```

## Gráfico "do jeito que veio ao mundo"
### Primeiro, iremos plotar um gráfico e deixando apenas as configurações padrão.

```python
fig, ax = plt.subplots(figsize=(9,6), dpi=72)

#plot das linhas
ax.plot(range(12), alunos, color=CINZA)
ax.plot(range(12), projetos, color=AZUL)

plt.show()
```

## AJUSTE 01
### Nesse primeiro momento, removeremos os spines da direita e superior, além de "esticar" um pouco o gráfico para uma visão geral melhor.

```python
fig, ax = plt.subplots(figsize=(9,6), dpi=72)

#plot das linhas
ax.plot(range(12), alunos, color=CINZA)
ax.plot(range(12), projetos, color=AZUL)

#Removendo Spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#Mudando os limites do eixo Y
ax.set_ylim(0, 300)

plt.show()
```

## Ajuste 2
### Agora, começaremos a estilizar de fato nosso gráfico. Iremos trazer uma espessura maior às linhas do gráfico. Além disso, configuraremos os ticks dos eixos X e Y, removendo os do eixo Y, e configurando e renomeando os ticks do eixo X com os meses do ano.

```python
fig, ax = plt.subplots(figsize=(9,6), dpi=72)

#plot das linhas
ax.plot(range(12), alunos, color=CINZA, linewidth=4)
ax.plot(range(12), projetos, color=AZUL, linewidth=4)

#Removendo Spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#Mudando os limites do eixo Y
ax.set_ylim(0, 300)

#Remover yticks
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

#Configurar xticks
ax.set_xticks(ticks=range(12))

#Nomear xticks
ax.set_xticklabels(X)

plt.show()
```

## Ajuste 3
### Para trazer uma clareza maior, iremos deixar visíveis os 3 últimos valores no estilo de scatters plots. Além disso, configuraremos seus tamanhos e cores.

```python
fig, ax = plt.subplots(figsize=(9,6), dpi=72)

#plot das linhas
ax.plot(range(12), alunos, color=CINZA, linewidth=4)
ax.plot(range(12), projetos, color=AZUL, linewidth=4)

#Removendo Spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#Mudando os limites do eixo Y
ax.set_ylim(0, 300)

#Remover yticks
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

#Configurar xticks
ax.set_xticks(ticks=range(12))

#Nomear xticks
ax.set_xticklabels(X)

#Plotando os scatters plots
ax.scatter(range(12)[-3:], alunos[-3:], s=90, color=CINZA)
ax.scatter(range(12)[-3:], projetos[-3:], s=90, color=AZUL)

#Anotações nos pontos de scatters
for i, j in enumerate(alunos[-3:]):
  ax.annotate(
      j,
      xy=(i+9, j),
      xytext=(5, 7),
      textcoords='offset points',
      color=CINZA_TEXTO,
      fontsize=12,
      fontweight='medium'

  )

for i, j in enumerate(projetos[-3:]):
  ax.annotate(
      j,
      xy=(i+9, j),
      xytext=(5,7),
      textcoords='offset points',
      color=AZUL,
      fontsize=12,
      fontweight='medium'
  )

plt.show()
```

## Ajuste Final
### Por fim, vamos inserir os textos necessários para um entendimento completo do gráfico e do que ele quer transmitir. Traremos o título, subtítulo e rodapé. Não só isso, incluiremos os nomes de cada linha do gráfico, de um modo clean e que não "suje" a vista de quem está lendo.

```python
print("\n\n\n\n")

fig, ax = plt.subplots(figsize=(10,6), dpi=72)

#plot das linhas
ax.plot(range(12), alunos, color=CINZA, linewidth=4)
ax.plot(range(12), projetos, color=AZUL, linewidth=4)

#Removendo Spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#Mudando os limites do eixo Y
ax.set_ylim(100, 220)

#Remover yticks
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

#Configurar xticks
ax.set_xticks(ticks=range(12))

#Nomear xticks
ax.set_xticklabels(X)

#Plotando os scatters plots
ax.scatter(range(12)[-3:], alunos[-3:], s=90, color=CINZA)
ax.scatter(range(12)[-3:], projetos[-3:], s=90, color=AZUL)

#Anotações nos pontos de scatters
for i, j in enumerate(alunos[-3:]):
  ax.annotate(
      j,
      xy=(i+9, j),
      xytext=(5, 7),
      textcoords='offset points',
      color=CINZA_TEXTO,
      fontsize=12,
      fontweight='bold'

  )

for i, j in enumerate(projetos[-3:]):
  ax.annotate(
      j,
      xy=(i+9, j),
      xytext=(5,7),
      textcoords='offset points',
      color=AZUL,
      fontsize=12,
      fontweight='bold'
  )

#Legenda no final de cada linha
ax.annotate(
    "Alunos",
    xy=(len(alunos)-1, alunos[-1]),
    xytext=(50,0),
    textcoords='offset points',
    color=CINZA,
    fontsize=18,
    fontweight='semibold'
)

ax.annotate(
    "Projetos",
    xy=(len(projetos)-1, projetos[-1]),
    xytext=(50,0),
    textcoords='offset points',
    color=AZUL,
    fontsize=18,
    fontweight='semibold'
)

#Definindo o título e subtítulo
ax.text(-0.5, 250, 'Taxa de Engajamento', fontsize=24, color=CINZA_TEXTO, fontweight='bold')
ax.text(-0.5, 230, 'KPI de engajamento, que considera a quantidade\nde acessos únicos e projetos enviados para\nrevisão', fontsize=13, color=CINZA)

#Rodapé
ax.text(
    -0.5,80,
    "Fonte: Dados extraídos a partir da Hotmart e Memberkit (21/06/2021)",
    fontsize=12,
    color=CINZA
)


plt.show()
plt.savefig('grafico.png', format='png')
```
